```python
remain_demand = pd.read_csv('G:/내 드라이브/ESC-22FALL/빅콘테스트/preprocessed data/remain_demand_data.csv' ,encoding='cp949')
```

## 법정구역별 전기차 등록 계수 적용 

```python
adng_car_registration = pd.read_csv('G:/내 드라이브/ESC-22FALL/빅콘테스트/Preprocessed Data/adng_car_registration.csv', encoding='cp949')
adng_car_registration
```

```python
import re

# XX1동과 같이 숫자가 포함된 행정동에서 숫자 제거
def adng_combine(adng_nm):
    new_adng_nm = re.sub(r"[0-9]", "", adng_nm)
    return new_adng_nm
```

```python
# 숫자 제거한 법정읍면동 정보를 adng_nm2컬럼에 저장
remain_demand['adng_nm2'] = remain_demand.apply(lambda row: adng_combine(row['adng_nm']), axis=1)
```

```python
# 생성한 전기차 비율 계수를 아파트/비아파트 데이터 count_cust에 적용
coef_dict = {}
for i in adng_car_registration.index:
    coef_dict[adng_car_registration.loc[i, 'adng_nm']] = adng_car_registration.loc[i, 'ratio_coef']
# 전기차 비율 계수가 적용된 adj_count_cust
remain_demand['adj_count_cust'] = remain_demand.apply(lambda row: row['count_cust']*coef_dict[row['adng_nm2']], axis=1)
remain_demand
```

## 법정 구역별 클러스터링

```python
total_remaind_demand = remain_demand['adj_count_cust'].sum()                            # 남은 수요점 총 count_cust 합 계산
adng_remain_demand = remain_demand.groupby('adng_nm2')['adj_count_cust'].sum()          # 법정 구역별 수요점 counst_cust 합 계산
adng_remain_demand_ratio = adng_remain_demand/total_remaind_demand                      # 법정 구역별 수요점 ratio
# 목표 달성을 위해 1500대의 충전기 추가 필요 
# 1500대? <--- 현재 notapart충전소 충전기수 = 342(완속충전기) + 237(급속충전기)*4 = 1290 가 46%충족 we need 54% more!
adng_goal_charger_num = pd.DataFrame(np.round(adng_remain_demand_ratio*1500, 0)).reset_index()  # 1500대를 각 법정읍면동의 수요 총합 비율로 배분               
adng_goal_charger_num.head(5)
```

```python
remain_demand['cluster_label'] = 0  #remain_demand데이터에 'cluster_label'컬럼 추가. 후에 클러스터링 이후 해당 컬럼에 군집 label저장

# start with '구갈동'
adng_demand = remain_demand[remain_demand['adng_nm2'] == '구갈동']  # remian_demand에서 ['adng_nm2'] == '구갈동'인 데이터 추출
n_clusters = adng_goal_charger_num.loc[0, 'adj_count_cust']         # 군집 개수를 법정읍면동별 목표 충전기 대수로 설정

Kmeans = KMeans(n_clusters=int(n_clusters))                         # 군집 개수가 n_cluster인 kmeans객체 생성
Kmeans.fit_transform(adng_demand[['cell_xcrd', 'cell_ycrd']])       # 수요점 데이터 학습

# 수요점 데이터에 cluster label 저장
adng_demand['cluster_label'] = Kmeans.labels_   # 클러스트링 결과 각 수요점 별 cluster_label 추출
for i in adng_demand.index:                     # remain_demand에 cluster_label 저장
    remain_demand.loc[i, 'cluster_label'] = adng_demand.loc[i, 'cluster_label']

# 군집점 데이터 저장
clusters = pd.DataFrame(Kmeans.cluster_centers_, columns=['cell_xcrd', 'cell_ycrd'])                        # 군집점 x,y 좌표 저장
clusters['count_point'] = pd.Series(adng_demand.groupby('cluster_label')['adj_count_cust'].count())         # 군집점 별 수요점 개수 저장
clusters['adj_count_cust_sum'] = pd.Series(adng_demand.groupby('cluster_label')['adj_count_cust'].sum())    # 군집점 별 수요점 adj_count_cust 총합 저장
# 군집점 별 충전기 대수 = 법정읍면동별 충전기 대수 * (군집점 adj_count_cust 합/법정읍면동별 adj_count_cust 합)
clusters['charger_number'] = np.round(clusters['adj_count_cust_sum']/clusters['adj_count_cust_sum'].sum() * 154, 0) 
clusters.reset_index(drop=False, inplace=True)                      # 군집점 데이터 인덱스 초기화, 기존 인덱스 저장(drop=False)
clusters.rename(columns={'index':'cluster_label'}, inplace=True)    # 기존 인덱스가 cluster_label의 역할 <- 
clusters['adng_nm'] = '구갈동'                                      # 법정읍면동 이름 저장
```

```python
clusters
```

```python
# 앞서 구갈동 데이터에만 진행했던 클러스터링 모두 진행. 과정은 거의 동일
for i in adng_goal_charger_num[1:].index:
    adng_demand = remain_demand[remain_demand['adng_nm2'] == adng_goal_charger_num.loc[i, 'adng_nm2']]
    n_clusters = adng_goal_charger_num.loc[i, 'adj_count_cust']

    Kmeans = KMeans(n_clusters=int(n_clusters))
    Kmeans.fit_transform(adng_demand[['cell_xcrd', 'cell_ycrd']])

    adng_demand['cluster_label'] = Kmeans.labels_
    for index in adng_demand.index:
        remain_demand.loc[index, 'cluster_label'] = adng_demand.loc[index, 'cluster_label']

    clusters_i = pd.DataFrame(Kmeans.cluster_centers_, columns=['cell_xcrd', 'cell_ycrd'])
    clusters_i['count_point'] = pd.Series(adng_demand.groupby('cluster_label')['adj_count_cust'].count())
    clusters_i['adj_count_cust_sum'] = pd.Series(adng_demand.groupby('cluster_label')['adj_count_cust'].sum())
    clusters_i['charger_number'] = np.round(clusters_i['adj_count_cust_sum']/clusters_i['adj_count_cust_sum'].sum() * adng_goal_charger_num.loc[i, 'adj_count_cust'], 0)
    clusters_i.reset_index(drop=False, inplace=True)
    clusters_i.rename(columns={'index':'cluster_label'}, inplace=True)
    clusters_i['adng_nm'] = adng_goal_charger_num.loc[i, 'adng_nm2']
    # 생성한 군집점 데이터를 앞서 진행한 군집점 데이터에 concatenate
    clusters = pd.concat([clusters, clusters_i])
    clusters.reset_index(drop=True, inplace=True)
```

```python
zero_charger = clusters[clusters['charger_number'] < 1]     # 군집점에 충전기 대수가 2대 미만이면 해당 군집점 삭제
clusters_final = clusters.drop(zero_charger.index, axis=0)
clusters_final['charger_number'].sum()  # 줄어든 총 충전기 대수 확인
```

```python
clusters_final.reset_index(drop=True, inplace=True)
cluster_final = clusters_final[['adng_nm', 'cluster_label', 'cell_xcrd', 'cell_ycrd', 'charger_number', 'count_point', 'adj_count_cust_sum']]
cluster_final
```
