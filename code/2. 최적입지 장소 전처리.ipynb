```python
import requests, json

def get_location(address):
  url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + address
  # 'KaKaoAK '는 그대로 두시고 개인키만 지우고 입력해 주세요.
  # ex) KakaoAK 6af8d4826f0e56c54bc794fa8a294
  headers = {"Authorization": "KakaoAK 0169c7c2241c662713182accff2350cf"}
  api_json = json.loads(str(requests.get(url,headers=headers).text))
  address = api_json['documents'][0]['address']
  crd = {"lat": str(address['y']), "lng": str(address['x'])}
  address_name = address['address_name']

  return crd

crd = get_location("제주 애월읍")
print(crd)
```

```python
#철자오류 영덕-덕영 
#수원에 위치한 자료 drop
#숫자 7이 아니라 1로 잘못 표기 변경
#인식이 안되는 자료가 있어서 지번으로 대체
```

```python
#distance함수는 두 좌표간 거리를 재는 함수입니다.
```

```python
def distance(index):
    place = (float(opt_sites[[0],[1]]), float(opt_sites[[0],[0]])) #optimal point
    tuple1 = (float(newarray[[index],[1]])),float(newarray[[index],[0]]) #거리를 재려는 위치
    #print(tuple1)
    dist = haversine(tuple1, place, unit = 'km')
    return round(dist,6)
```
